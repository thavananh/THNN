```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
aspects = [
    "Teaching quality",
    "Support from lectures",
    "Course Information",
    "Organization and management",
    "General review",
    "Learning environment",
    "Workload",
    "Test and evaluation"
]
positive = [4580, 3497, 632, 359, 696, 184, 242, 127]
neutral  = [540, 268, 541, 338, 448, 288, 243, 187]
negative = [1800, 863, 730, 755, 303, 508, 271, 346]

# Plot setup
x = np.arange(len(aspects))
fig, ax = plt.subplots(figsize=(20, 12))

# Colors matching previous
colors = {
    'Positive': '#4CAF50',  # Mu xanh l cy ti sng, tng trng cho tch cc
    'Negative': '#F44336',  # Mu  ti, ph hp cho tiu cc
    'Neutral':  '#FFC107'   # Mu vng cam, trung tnh, d nhn
}

# Stacked bars
ax.bar(x, positive, label='Positive', color=colors['Positive'])
ax.bar(x, negative, bottom=positive, label='Negative', color=colors['Negative'])
bottom_neutral = [p + n for p, n in zip(positive, negative)]
ax.bar(x, neutral, bottom=bottom_neutral, label='Neutral', color=colors['Neutral'])

# Labels and legend
ax.set_xticks(x)
ax.set_xticklabels(aspects, rotation=45, ha='right')
ax.set_ylabel('Count')
ax.set_title('Sentiment Polarity by Aspect (Stacked)')
ax.legend()

plt.tight_layout()
plt.show()
```
